```python
import pandas as pd                        
```

```python
# Unofficial API for Google Trends
from pytrends import dailydata
```

```python
# get information on how often "party decorations" was searched on Google from 1/1/2020 to 12/31/2020
df = dailydata.get_daily_data('party decorations', 2020, 1, 2020, 12, geo = 'US')
```

```python
df.info()
```

```python
df.describe()
```

```python
df.fillna(0, inplace=True)
```

```python
df.info()
```

```python
df = df.rename(columns={'party decorations_unscaled': 'unscaled',
                        'party decorations_monthly': 'monthly',
                        'party decorations': 'cat'})
df.tail()
```

```python
df.to_csv('data/party_decorations_API.csv')
```
